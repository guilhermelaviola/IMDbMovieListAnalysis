```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter
sns.set(style = 'darkgrid')
plt.style.use('seaborn-pastel')
```

```python
actors = pd.read_excel('My favorites.xlsx')
actors.head()
```

```python
actors.shape
```

```python
actors['Birth Day and Month'] = actors['Birth Day and Month'].map(pd.to_datetime)
actors['Day'] = actors['Birth Day and Month'].apply(lambda x : x.day)
actors['Month'] = actors['Birth Day and Month'].apply(lambda x : x.month)
actors['Year'] = actors['Birth Day and Month'].apply(lambda x : x.year)

print(actors.head())
```

```python
# Displaying actors by birth year:
actors.groupby('Birth Day and Month').first()['Birth Year'].value_counts().plot(kind = 'pie',
                                                              autopct = '%.0f%%',
                                                              figsize = (8, 8),
                                                              wedgeprops = dict(width = 0.4),
                                                              pctdistance = 0.8)
plt.ylabel(None)
plt.title('Breakdown of actors by Birth Year', fontweight = 'bold')
plt.show()
```

```python
# Displaying actors by birth month:
actors.groupby('Birth Day and Month').first()['Month'].value_counts().plot(kind = 'pie',
                                                              autopct = '%.0f%%',
                                                              figsize = (8, 8),
                                                              wedgeprops = dict(width = 0.4),
                                                              pctdistance = 0.8)
plt.ylabel(None)
plt.title('Breakdown of actors by Birth Month', fontweight = 'bold')
plt.show()
```
