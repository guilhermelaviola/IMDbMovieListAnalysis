```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter
sns.set(style = 'darkgrid')
plt.style.use('seaborn-pastel')
```

```python
# Importing the dataset and displaying the last 10 rows:
actors = pd.read_excel('My favorites.xlsx')
actors.tail()
```

```python
# Displaying the number of rows and columns in the dataset:
actors.shape
```

```python
# Creating a datetime object based on the dates from the dataset:
actors['Birth Day and Month'] = actors['Birth Day and Month'].map(pd.to_datetime)
actors['Day'] = actors['Birth Day and Month'].apply(lambda x : x.day)
actors['Month'] = actors['Birth Day and Month'].apply(lambda x : x.month)
actors['Year'] = actors['Birth Day and Month'].apply(lambda x : x.year)
actors['Weekday'] = actors['Birth Day and Month'].apply(lambda x : x.weekday())
```

```python
# Displaying the last 10 rows after the datetime columns were added:
actors.tail()
```

```python
# Displaying the months with most and less birthdays:
actors['Month'].value_counts().plot(kind = 'barh', color = 'yellow', figsize = (8, 4))
plt.ylabel(None)
plt.title('Birthdays by month', fontweight = 'bold')
plt.show()
```

```python
# Displaying the years with most and less birthdays:
actors['Year'].value_counts().plot(kind = 'barh', color = 'yellow', figsize = (8, 16))
plt.ylabel(None)
plt.title('Birthdays by year', fontweight = 'bold')
plt.show()
```

```python
# Displaying the days of the month with most and less birthdays:
actors['Day'].value_counts().plot(kind = 'barh', color = 'yellow', figsize = (8, 12))
plt.ylabel(None)
plt.title('Birthdays by days of the month', fontweight = 'bold')
plt.show()
```

```python
# Displaying the days of the week with most and less birthdays:
actors['Weekday'].value_counts().plot(kind = 'barh', color = 'yellow', figsize = (8, 4))
plt.ylabel(None)
plt.title('Birthdays by days of the week', fontweight = 'bold')
plt.show()
```

```python
# Exporting the dataset:
actors.to_csv('out.csv', index=False)
```
